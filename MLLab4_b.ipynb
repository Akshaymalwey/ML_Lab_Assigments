```python
def transpose(matrix):
    if not matrix:
        return []
    rows = len(matrix)
    cols = len(matrix[0])
    transposed_matrix = [[0 for _ in range(rows)] for _ in range(cols)]
    for i in range(rows):
        for j in range(cols):
            transposed_matrix[j][i] = matrix[i][j]
    return transposed_matrix

def multiply_matrices(matrix_a, matrix_b):
    if not matrix_a or not matrix_b:
        return []

    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0])
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0])

    if cols_a != rows_b:
        raise ValueError("Error")

    result_matrix = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]
    return result_matrix

def determinant_2x2(matrix):
    if len(matrix) != 2 or len(matrix[0]) != 2 or len(matrix[1]) != 2:
        raise ValueError("Matrix must be 2x2.")
    return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]

def inverse_2x2(matrix):
    det = determinant_2x2(matrix)
    if det == 0:
        raise ValueError("Cannot invert.")

    return [
        [matrix[1][1] / det, -matrix[0][1] / det],
        [-matrix[1][0] / det, matrix[0][0] / det]
    ]

def general_inverse_placeholder(matrix):
    raise NotImplementedError("Use specialized libraries for this.")
```

```python
def generate_polynomial_features(X_raw):
    if not X_raw:
        return []

    design_matrix = []
    for sample in X_raw:
        processed_sample = [1.0]
        for feature_val in sample:
            processed_sample.append(feature_val + feature_val**2)
        design_matrix.append(processed_sample)

    return design_matrix

def solve_polynomial_regression(X_raw, y_target):
    y_matrix = [[val] for val in y_target]
    X_poly = generate_polynomial_features(X_raw)
    X_poly_T = transpose(X_poly)
    X_poly_T_X_poly = multiply_matrices(X_poly_T, X_poly)
    matrix_dim = len(X_poly_T_X_poly)
    if matrix_dim == 2:
        try:
            X_poly_T_X_poly_inv = inverse_2x2(X_poly_T_X_poly)
        except ValueError as e:
            print(f"Error during inversion: {e}")
            return None
    elif matrix_dim == 1:
        if X_poly_T_X_poly[0][0] == 0:
             print("Matrix is singular, cannot invert.")
             return None
        X_poly_T_X_poly_inv = [[1 / X_poly_T_X_poly[0][0]]]
    else:
        print("Warning")
        return None
    X_poly_T_y = multiply_matrices(X_poly_T, y_matrix)
    weights_matrix = multiply_matrices(X_poly_T_X_poly_inv, X_poly_T_y)
    weights = [w[0] for w in weights_matrix]
    return weights

def predict_polynomial_regression(X_raw, weights):
    if not weights or not X_raw:
        return []

    predictions = []
    for sample in X_raw:

        processed_sample_features = [1.0]
        for feature_val in sample:
            processed_sample_features.append(feature_val + feature_val**2)

        prediction = 0
        for i in range(len(weights)):
            prediction += weights[i] * processed_sample_features[i]
        predictions.append(prediction)
    return predictions
```
